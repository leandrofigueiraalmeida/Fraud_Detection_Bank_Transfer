# 1° What's Business Problem

This fictitious dataset has transfer transactions between accounts of the same Financial Institution. There are more than 200,000 records of transfers, 490 of which were fraudulent, that is, they were not carried out by the owner of the checking account. We have 28 explanatory variables and the Class variable is the Target and takes the value 1 in case of fraud and 0 otherwise.

# 3° Data Preparation

* Split Dataframe into training and test Dataset

```python
# Separando os dados para treinamento e teste da máquina preditiva
X = df.drop(['Target'], axis=1)
y = df['Target']
```

* SMOTE

```python
#Balanceamento da Classe (Target)
smt = SMOTE()
```

* Balance Data

```python
X, y = smt.fit_resample(X, y)
```

* Count Data to check SMOTE balance

```python
#Contagem do target depois do smote. Obs: Antes da contagem, para usar a função bincount, precisei converter de float para inteiro
yi = y.astype(int)
np.bincount(yi)
```

```python
ax = sns.countplot(x=yi)
```

* Split Dataframe into training and test Dataset

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=7,stratify=y)
```

# 4° Predictive Machine to Fraud Detection

* Predictive Machine
* 3 Types of model to test: XGBoost / LightGBM / Random Forest and

#4.1 XGBoost

* Predictive Machine and fit the model to the data we prepare for training.

* Classify with the data we have prepared for evaluation.

* Evaluate model performance with key ranking metrics.

```python
# Construindo o Modelo
from xgboost import XGBClassifier
model = XGBClassifier()
```

* Training Model

```python
# Treinando o Modelo
model = model.fit(X_train, y_train)
model
```

```python
# Passando novos dados para o modelo
y_predict = model.predict(X_val)
```

## 5° Evaluating Predictive Machine - XGBoost

* Comparing Data and Forecasts

```python
# Comparando gabarito e Previsoes_da_Máquina
gabarito = pd.DataFrame({'Gabarito': y_val,'Previsoes_da_Máquina': y_predict})
gabarito
```

* Evaluate Model

```python
# Evaluate model
print('Classification metrics: \n', classification_report(y_val, y_predict))
print('Acurácia: \n', accuracy_score(y_val, y_predict))

print('Confusion Matrix: \n', confusion_matrix(y_val, y_predict))
```

#4.2 LightGBM

* Training Model

```python
# Light GBM
train_data=lgb.Dataset(X_train,label=y_train)
#setting parameters for lightgbm
param = {'num_leaves':1000,         #número de folhas em uma árvore; padrão = 31; type = int
         'objective':'binary',
         'max_depth':7,
         'learning_rate':.01,
         'max_bin':200}

param['metric'] = ['auc', 'binary_logloss']

#training our model using light gbm
num_round=50

lgbm=lgb.train(param,train_data,num_round)

```

```python
# Passando novos dados para o modelo
y_predict = lgbm.predict(X_val)
```

## 5° Evaluating Predictive Machine - LightGBM

* Comparing Data and Forecasts

```python
# Comparando gabarito e Previsoes_da_Máquina
gabarito = pd.DataFrame({'Gabarito': y_val,'Previsoes_da_Máquina': y_predict})
gabarito
```

```python
y_predict.size
```

* Converting probabilities into 0 or 1

```python
#converting probabilities into 0 or 1
for i in range(0,170589):
    if y_predict[i]>=.5:       # setting threshold to .5
       y_predict[i]=1
    else:
       y_predict[i]=0
```

* Evaluate Model

```python
# Evaluate model
print('Classification metrics: \n', classification_report(y_val, y_predict))
print('Acurácia: \n', accuracy_score(y_val, y_predict))
#0.9843  F1 98%
print('Confusion Matrix: \n', confusion_matrix(y_val, y_predict))
```
